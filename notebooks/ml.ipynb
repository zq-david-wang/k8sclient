```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# load the data
row = pd.read_excel("/home/xzy/PycharmProjects/misc/myproject/data/data.xls")
```

```python
print row.shape
# filter out na values
print row.dropna().shape
```

```python
# ingore columns[3,4]
new_columns = row.columns[:3].union(row.columns[5:])
xdata = row.dropna()[new_columns]
xdata
```

```python
# PCA for viewing the data as 2d
from sklearn import decomposition
pca = decomposition.PCA(n_components=2)
pca.fit(xdata)
X = pca.transform(xdata)
plt.scatter(X[:, 0], X[:, 1])
```

```python
# PCA for viewing the data as 3d
from mpl_toolkits.mplot3d import Axes3D
pca_3d = decomposition.PCA(n_components=3)
pca_3d.fit(xdata)
X = pca_3d.transform(xdata)
fig = plt.figure(1, figsize=(12, 8))
plt.clf()
ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134)
ax.scatter(X[:, 0], X[:, 1], X[:, 2])
```

```python
# kmeans pca reduced data -  then  plot
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import numpy as np

reduced_data = PCA(n_components=2).fit_transform(xdata)
kmeans = KMeans(init='k-means++', n_clusters=6, n_init=10)
kmeans.fit(reduced_data)
h = .02
x_min, x_max = reduced_data[:, 0].min() - 1, reduced_data[:, 0].max() + 1
y_min, y_max = reduced_data[:, 1].min() - 1, reduced_data[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.figure(1)
plt.clf()
plt.imshow(Z, interpolation='nearest',
           extent=(xx.min(), xx.max(), yy.min(), yy.max()),
           cmap=plt.cm.Paired,
           aspect='auto', origin='lower')
plt.scatter(reduced_data[:, 0], reduced_data[:, 1], color='w')
```

```python
# kmeans origin data and then color plot
kmeans = KMeans(init='k-means++', n_clusters=6, n_init=10)
kmeans.fit(xdata)
z = kmeans.predict(xdata)
plt.scatter(reduced_data[:, 0], reduced_data[:, 1], c=z)
```

```python
# MiniBatchKMeans origin data and then color plot
from sklearn.cluster import MiniBatchKMeans
minbatchkmeans = MiniBatchKMeans(init='k-means++', n_clusters=6, batch_size=6,
                      n_init=10, max_no_improvement=10, verbose=0)
minbatchkmeans.fit(xdata)
z = minbatchkmeans.predict(xdata)
plt.scatter(reduced_data[:, 0], reduced_data[:, 1], c=z)
```

```python
from sklearn.cluster import MeanShift
ms = MeanShift(bandwidth=6, bin_seeding=True)
ms.fit(X)
y = ms.labels_
plt.scatter(reduced_data[:, 0], reduced_data[:, 1], c=y)
```

```python
from sklearn.cluster import AgglomerativeClustering
ward = AgglomerativeClustering(n_clusters=6, linkage='ward')
ward.fit(X)
y = ward.labels_
plt.scatter(reduced_data[:, 0], reduced_data[:, 1], c=y)
```

```python
# 1d pca and kmeans
reduced_1d = PCA(n_components=1).fit_transform(xdata)
kmeans = KMeans(init='k-means++', n_clusters=6, n_init=30)
kmeans.fit(reduced_1d)
z = kmeans.predict(reduced_1d)
plt.scatter(reduced_data[:, 0], reduced_data[:, 1], c=z)
```
